```python
import numpy as np # linear algebra
import pandas as pd 
import matplotlib.pyplot as plt
```

### Load Data

```python
#data = pd.read_csv("https://covid.ourworldindata.org/data/owid-covid-data.csv")
#data.to_csv('Coronavirus_cases_daily_update.csv', encoding='utf-8', index=False)
data =  pd.read_csv('Coronavirus_cases_daily_update.csv', sep = ',')
data.head()
```

```python
data = data[['iso_code', 'total_cases']]
data.head()
```

### Data Preparation (Processing)

```python
x = np.array(data['iso_code']).reshape(-1,1)
y = np.array(data['total_cases']).reshape(-1,1)

print(data.isnull().sum())
data =  data.dropna() #drop null values

print(data.isnull().sum())

plt.plot(y)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

polyFeat = PolynomialFeatures(degree=2) 
x = polyFeat.fit_transform(x)
print(x)
```

### Training 

```python
from sklearn import linear_model

model = linear_model.LinearRegression()
model.fit(x,y)

accuracy = model.score(x,y)
print("accurary: ", round(accuracy, 2))
```

```python
y0 = model.predict(x)
plt.plot(y0, '--b')
plt.show()
```
